```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as mp

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost-revenue-clean.csv')
```

```python
data
```

```python
data.describe()
```

```python
X= DataFrame(data, columns=['production_budget_usd'])
y= DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
## Adjust the size of the chart
mp.figure(figsize=(10,6))
## Call the Scatterplot Functionality (Alpha sets the transparency of the dots)
mp.scatter(X,y, alpha = 0.3)
## Add a title
mp.title('Film Cost vs Global Revenue')
## Add a label to the x-axis
mp.xlabel('Production Budget $')
## Add a label to the y-axis
mp.ylabel('Worldwide Gross $')
## Set value limits to clean the visualization on y-axis
mp.ylim(0, 3000000000)
## Set value limits to clean the visualization on the x-axis
mp.xlim(0, 450000000)
## Show the scatterplot
mp.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

```python
## Theta_1
regression.coef_
```

```python
## Intercept
regression.intercept_
```

```python
## Adjust the size of the chart
mp.figure(figsize=(10,6))
## Call the Scatterplot Functionality (Alpha sets the transparency of the dots)
mp.scatter(X,y, alpha = 0.3)
## Plot Regression Model Prediction
mp.plot(X, regression.predict(X), color='red')
## Add a title
mp.title('Film Cost vs Global Revenue')
## Add a label to the x-axis
mp.xlabel('Production Budget $')
## Add a label to the y-axis
mp.ylabel('Worldwide Gross $')
## Set value limits to clean the visualization on y-axis
mp.ylim(0, 3000000000)
## Set value limits to clean the visualization on the x-axis
mp.xlim(0, 450000000)
## Show the scatterplot
mp.show()
```

```python
## Amount of variation in film revenue explained by Film's budget
## Returing r-squared value
regression.score(X,y)
```
